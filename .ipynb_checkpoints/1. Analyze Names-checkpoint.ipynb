# Preparation:

For the analysis, we will set up <a href="https://docs.deeppavlov.ai/en/master/features/models/NER.html">Deep Pavlov</a>.

Notice: As of May. 2023, due to issue with pytorch, deeppavlov only work with Python 3.10 or earlier versions.

```python
from deeppavlov import build_model

ner_model = build_model('ner_ontonotes_bert', download=True, install=True)
```

```python
tagged_tokens = ner_model(['Bob Ross lived in Florida', 'Elon Musk founded Tesla'])
tagged_tokens
```

```python
tokens = tagged_tokens[0]
tags = tagged_tokens[1]

for x , y in zip(tokens, tags):
    print(x)
    print(y)
    words_in_sentence = []
    for word in zip(x,y):
        words_in_sentence.append({
            'token': word[0],
            'type': word[1]
        })
    
    print(words_in_sentence)
        # print(word)
    # for i in range(len(x)):
    #     print(x[i], y[i])
    print("")
```
