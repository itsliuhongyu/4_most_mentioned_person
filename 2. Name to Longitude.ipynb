```python
import pandas as pd

df = pd.read_csv('namesigns.csv')
df
```

```python
pd.set_option('display.max_rows', None)
```

```python
robert_moses_rows = df[df['names'] == 'robert moses']
robert_moses_rows.head(10)
```

```python
# !pip install geopy
```

```python
robert_moses_rows['combined'] = robert_moses_rows['name'] + ", " + robert_moses_rows['borough']

robert_moses_rows
```

```python
from geopy.geocoders import get_geocoder_for_service
from geopy.geocoders import Nominatim

get_geocoder_for_service("nominatim")

geolocator = Nominatim(user_agent="placesearcher")

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)



# Code while using pandas
robert_moses_rows['location'] = robert_moses_rows['combined'].apply(geocode)

robert_moses_rows['latitude'] = robert_moses_rows['location'].apply(lambda loc: tuple([loc.latitude])[0] if loc is not None else None)
robert_moses_rows['longitude'] = robert_moses_rows['location'].apply(lambda loc: tuple([loc.longitude])[0] if loc is not None else None)

```

```python
robert_moses_rows.to_csv('robert_moses_rows.csv', index=False)
```
